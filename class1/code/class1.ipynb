```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 6, 8, 10])

# Hyperparameters
lr = 0.01         # Learning rate
epochs = 1000     # Number of iterations

# Initialize parameters
m = 0  # Slope
c = 0  # Intercept
n = len(X)

# Gradient Descent loop
for i in range(1, epochs + 1):
    Y_pred = m * X + c
    error = Y - Y_pred

    # Compute gradients
    dm = (-2 / n) * np.sum(X * error)
    dc = (-2 / n) * np.sum(error)

    # Update parameters
    m -= lr * dm
    c -= lr * dc

    # Compute loss (Mean Squared Error)
    loss = np.mean(error ** 2)

    # ✅ Print every iteration
    print(f"Step {i:>4}: m = {m:.4f}, c = {c:.4f}, Loss = {loss:.6f}")

# Final result
print(f"\nFinal model: y = {m:.2f}x + {c:.2f}")

# Plotting
plt.scatter(X, Y, label="Actual Points")
plt.plot(X, m * X + c, color='red', label="Fitted Line")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gradient Descent Linear Regression")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss function: L(w) = (w - 3)^2 + 2
def loss(w):
    return (w - 3)**2 + 2

# Define the gradient of the loss function: dL/dw = 2(w - 3)
def gradient(w):
    return 2 * (w - 3)

# Initialize parameters
w = 0  # start far from the minimum
learning_rate = 0.1
iterations = 10

# Store values for plotting
w_values = [w]
loss_values = [loss(w)]

# Gradient Descent Loop
for i in range(iterations):
    grad = gradient(w)
    w = w - learning_rate * grad
    w_values.append(w)
    loss_values.append(loss(w))

# Plotting the loss curve and the steps
w_range = np.linspace(-1, 7, 200)
loss_range = loss(w_range)

plt.figure(figsize=(10, 6))
plt.plot(w_range, loss_range, label='Loss Function')
plt.scatter(w_values, loss_values, color='red', label='Gradient Descent Steps')
for i in range(len(w_values)-1):
    plt.arrow(w_values[i], loss_values[i], 
              w_values[i+1]-w_values[i], loss_values[i+1]-loss_values[i],
              head_width=0.1, color='red', length_includes_head=True)
plt.title('Gradient Descent: Moving Opposite to Gradient')
plt.xlabel('Parameter (w)')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

# Generate synthetic nonlinear data
np.random.seed(42)
X = np.sort(np.random.rand(40, 1) * 2 - 1, axis=0)  # X in [-1, 1]
y = np.sin(1.5 * np.pi * X).ravel() + np.random.normal(0, 0.1, X.shape[0])  # nonlinear + noise

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

# Create models
model_no_reg = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
model_ridge = make_pipeline(PolynomialFeatures(degree=15), Ridge(alpha=1.0))

# Fit models
model_no_reg.fit(X_train, y_train)
model_ridge.fit(X_train, y_train)

# Predict
X_plot = np.linspace(-1, 1, 100).reshape(-1, 1)
y_pred_no_reg = model_no_reg.predict(X_plot)
y_pred_ridge = model_ridge.predict(X_plot)

# Plot results
plt.figure(figsize=(12, 5))

plt.scatter(X_train, y_train, color='black', label='Training Data')
plt.plot(X_plot, y_pred_no_reg, color='red', label='No Regularization')
plt.plot(X_plot, y_pred_ridge, color='green', linestyle='--', label='Ridge Regularization')
plt.title("Polynomial Regression (degree=15) with and without Regularization")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Create a synthetic dataset for linear regression
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add x0 = 1 to each instance for bias term
X_b = np.c_[np.ones((100, 1)), X]  # shape: (100, 2)

# Loss functions with and without regularization
def compute_loss(theta, X, y, l2_lambda=0):
    m = len(y)
    predictions = X.dot(theta)
    mse = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    l2_penalty = l2_lambda * np.sum(theta[1:] ** 2)
    return mse + l2_penalty

# Gradient of the cost function
def compute_gradient(theta, X, y, l2_lambda=0):
    m = len(y)
    predictions = X.dot(theta)
    grad = (1 / m) * X.T.dot(predictions - y)
    grad[1:] += (2 * l2_lambda) * theta[1:]  # L2 regularization term (excluding bias)
    return grad

# Gradient Descent Function
def gradient_descent(X, y, l2_lambda=0, learning_rate=0.1, n_iterations=50):
    theta = np.random.randn(2, 1)
    history = [theta.copy()]
    for _ in range(n_iterations):
        gradients = compute_gradient(theta, X, y, l2_lambda)
        theta = theta - learning_rate * gradients
        history.append(theta.copy())
    return np.array(history)

# Perform gradient descent without and with L2 regularization
history_no_reg = gradient_descent(X_b, y, l2_lambda=0)
history_l2_reg = gradient_descent(X_b, y, l2_lambda=10)

# Prepare the animation
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 2)
ax.set_ylim(0, 14)
ax.scatter(X, y, color='black', label='Training data')
line1, = ax.plot([], [], 'r-', label='No Regularization')
line2, = ax.plot([], [], 'g--', label='L2 Regularization')
ax.legend()

def init():
    line1.set_data([], [])
    line2.set_data([], [])
    return line1, line2

def animate(i):
    theta1 = history_no_reg[i]
    theta2 = history_l2_reg[i]
    x_vals = np.array([[1, x] for x in np.linspace(0, 2, 100)])
    y_vals1 = x_vals.dot(theta1)
    y_vals2 = x_vals.dot(theta2)
    line1.set_data(np.linspace(0, 2, 100), y_vals1)
    line2.set_data(np.linspace(0, 2, 100), y_vals2)
    return line1, line2

ani = animation.FuncAnimation(fig, animate, frames=len(history_no_reg), init_func=init,
                              blit=True, interval=200, repeat=False)

plt.close()  # Prevent static display

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate training updates for a single weight with L1 and L2 regularization
iterations = 50
eta = 0.1  # learning rate
lambda_l1 = 0.1
lambda_l2 = 0.1

# Initialize weight
w_l1 = 2.0
w_l2 = 2.0

w_l1_history = [w_l1]
w_l2_history = [w_l2]

# Perform updates
for _ in range(iterations):
    # Simulated gradient (let's assume gradient from data is always 1)
    grad = 1.0

    # L1 update (sign-based)
    grad_l1 = grad + lambda_l1 * np.sign(w_l1)
    w_l1 -= eta * grad_l1
    w_l1_history.append(w_l1)

    # L2 update (proportional to weight)
    grad_l2 = grad + lambda_l2 * w_l2
    w_l2 -= eta * grad_l2
    w_l2_history.append(w_l2)

# Plot the weight decay over iterations
plt.figure(figsize=(10, 6))
plt.plot(w_l1_history, label='L1 Regularization (Lasso)', linestyle='--', marker='o')
plt.plot(w_l2_history, label='L2 Regularization (Ridge)', linestyle='-', marker='s')
plt.title("Weight Update Dynamics with L1 vs L2 Regularization")
plt.xlabel("Iteration")
plt.ylabel("Weight Value")
plt.axhline(0, color='gray', linestyle=':')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Simulate training updates: No regularization vs L1 vs L2

# Reinitialize weights
w_no_reg = 2.0
w_l1 = 2.0
w_l2 = 2.0

w_no_reg_history = [w_no_reg]
w_l1_history = [w_l1]
w_l2_history = [w_l2]

# Perform updates
for _ in range(iterations):
    grad = 1.0

    # No Regularization
    w_no_reg -= eta * grad
    w_no_reg_history.append(w_no_reg)

    # L1 Regularization
    grad_l1 = grad + lambda_l1 * np.sign(w_l1)
    w_l1 -= eta * grad_l1
    w_l1_history.append(w_l1)

    # L2 Regularization
    grad_l2 = grad + lambda_l2 * w_l2
    w_l2 -= eta * grad_l2
    w_l2_history.append(w_l2)

# Plot the comparison
plt.figure(figsize=(10, 6))
plt.plot(w_no_reg_history, label='No Regularization', linestyle='-', marker='^')
plt.plot(w_l1_history, label='L1 Regularization (Lasso)', linestyle='--', marker='o')
plt.plot(w_l2_history, label='L2 Regularization (Ridge)', linestyle='-', marker='s')
plt.title("Weight Update Dynamics: No Reg vs L1 vs L2")
plt.xlabel("Iteration")
plt.ylabel("Weight Value")
plt.axhline(0, color='gray', linestyle=':')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
