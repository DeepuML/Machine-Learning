```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
import seaborn as sns 
import joblib
from sklearn.model_selection import train_test_split 
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
x, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=
                  30)
```

```python
x, y
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:, 0], x[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter plot of x')
plt.show()
```

```python
k_means_model = KMeans(n_clusters=4, init="k-means++",random_state=42)
```

```python
prediction = k_means_model.fit_predict(x)
```

```python
prediction
```

```python
centroids = k_means_model.cluster_centers_
```

```python
plt.scatter(x[:, 0], x[:, 1], c=prediction)
plt.scatter(centroids[:,0],centroids[:,1],color='red')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Cluster Assignments')
plt.show()
```

```python
inertia = []
k = range(1,10)
for n in k:
    km = KMeans(n_clusters=n ,init='k-means++')
    km.fit(x)
    inertia.append(km.inertia_)
    
```

```python
inertia
```

```python
plt.plot(k, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
elbow_idx = 3 
plt.plot(k[elbow_idx], inertia[elbow_idx], 'ro', markersize=10, label='Elbow Point (k=4)')
plt.legend()
plt.show()
```

```python
silhouette_score(x, prediction)
```

```python
joblib.dump(k_means_model, "K_Means_model.pkl")
```
