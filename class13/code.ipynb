```python
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
x
```

```python
x = x.values
```

```python
plt.scatter(x[:,0] , x[:,1])
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
plt.scatter(x_scaled[:,0] , x_scaled[:,1])
```

```python
dbscan_model1 = DBSCAN(eps=0.5, min_samples=4)
```

```python
label1 = dbscan_model1.fit_predict(x_scaled)
```

```python
plt.scatter(x_scaled[:,0] , x_scaled[:,1] , c= label1) 
```

```python
unique_labels = set(label1) - {-1}
centroids = np.array([x_scaled[label1 == k].mean(axis=0) for k in unique_labels])
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=label1, cmap='viridis', alpha=0.5)
for centroid in centroids:
    plt.scatter(*centroid, c='red', s=200, marker='x')
    circle = plt.Circle(centroid, 0.2, color='red', fill=False, linewidth=2)
    plt.gca().add_patch(circle)
plt.xlabel('Annual Income (scaled)')
plt.ylabel('Spending Score (scaled)')
plt.title('DBSCAN Clusters and Centroids')
plt.show()
centroids
```

```python
from sklearn.metrics import silhouette_score

eps_values = np.arange(0.1, 2.0, 0.05)
best_score = -1
best_eps = None

for eps in eps_values:
    db = DBSCAN(eps=eps, min_samples=4)
    labels_tmp = db.fit_predict(x_scaled)
    # Ignore cases where all points are noise or only one cluster
    if len(set(labels_tmp)) > 1 and len(set(labels_tmp) - {-1}) > 1:
        score = silhouette_score(x_scaled, labels_tmp)
        if score > best_score:
            best_score = score
            best_eps = eps

print(f"Best eps: {best_eps}, Best silhouette score: {best_score}")
```
