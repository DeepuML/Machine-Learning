Logistic Regression 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
import joblib
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
data
```

```python
data.target
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df
```

```python
df['target']=data.target
```

```python
df
```

```python
df['target_names'] = df['target'].map({i: name for i, name in enumerate(data.target_names)})
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

Binary Class Classifiaction using the Logistic Regerssion 

```python
binary_df=df[df['target']!=2]
```

```python
binary_df
```

```python
x_binary = binary_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
```

```python
y_binary = binary_df['target']
```

```python
x_test_binary, x_train_binary, y_test_binary, y_train_binary = train_test_split(x_binary,y_binary,test_size=0.2)
```

```python
logistic_binary = LogisticRegression()
logistic_binary.fit(x_train_binary,y_train_binary)
```

```python
prediction = logistic_binary.predict(x_test_binary)
```

```python
print("Classification report for the bianry class classification of logistic regression: \n", classification_report(y_test_binary, prediction))
```

```python
joblib.dump(logistic_binary,"Logistic_Regression_binary.pkl")
```

```python
accuracy_score(y_test_binary,prediction)
```

```python
confusion_matrix(y_test_binary,prediction)
```

Multi Class Classification using Logistic Regression

```python
x=df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
```

```python
y=df['target']
```

```python
x_train_multi, x_test_multi, y_train_multi, y_test_multi = train_test_split( x , y , test_size = 0.2)
```

```python
logistic_multi = LogisticRegression(multi_class='ovr', solver="lbfgs", max_iter=2000)
```

```python
logistic_multi.fit(x_train_multi,y_train_multi)
```

```python
prediction = logistic_multi.predict(x_test_multi)
```

```python
prediction
```

```python
print("Clasification report on multi class classifiaction using logistic regression : \n" , classification_report(y_test_multi, prediction))
```

```python
accuracy_score(y_test_multi,prediction)
```

```python
logistic_multi2 = LogisticRegression(multi_class='multinomial', solver="lbfgs", max_iter=2000)
```

```python
logistic_multi2.fit(x_train_multi, y_train_multi)
```

```python
prediction = logistic_multi2.predict(x_test_multi)
```

```python
prediction
```

```python
print("Clasification report on Multinomial Logistic Regression : \n", classification_report(y_test_multi,prediction))
```

```python
accuracy_score(y_test_multi,prediction)
```

```python
joblib.dump(logistic_multi,"Logistic_Regression_Multi.pkl")
```
